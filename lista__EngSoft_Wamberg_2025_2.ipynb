```python
# Questão 14 Solicita a média mínima para aprovação
media_minima = float(input("Digite a média mínima para aprovação: "))

# Solicita os dados do aluno
nome = input("Digite o nome do aluno: ")
sexo = input("Digite o sexo do aluno (M/F): ").strip().upper()

# Solicita as 3 notas
nota1 = float(input("Digite a primeira nota: "))
nota2 = float(input("Digite a segunda nota: "))
nota3 = float(input("Digite a terceira nota: "))

# Calcula a média
media = (nota1 + nota2 + nota3) / 3

# Determina se foi aprovado ou reprovado
aprovado = media >= media_minima

# Escolhe o artigo e a forma correta para "aprovado"
if sexo == 'M':
    artigo = "O aluno"
    status = "aprovado"
elif sexo == 'F':
    artigo = "A aluna"
    status = "aprovada"
else:
    artigo = "O(a) aluno(a)"
    status = "aprovado(a)"  # neutro para entrada inválida ou não-binária

# Exibe o resultado
if aprovado:
    print(f"{artigo} {nome} foi {status} com média {media:.2f}")
else:
    status = "reprovado" if sexo == 'M' else "reprovada"
    if sexo not in ['M', 'F']:
        status = "reprovado(a)"
    print(f"{artigo} {nome} foi {status} com média {media:.2f}")
```

```python
# 12
n = int(input("Quantas pessoas? "))
mais_velho_nome = ""
mais_velho_idade = -1
for _ in range(n):
    nome = input("Nome: ")
    idade = int(input("Idade: "))
    if idade > mais_velho_idade:
        mais_velho_nome = nome
        mais_velho_idade = idade
print("Pessoa mais velha:", mais_velho_nome)
```

```python
# 13

n = int(input("Quantas pessoas? "))
soma_m = soma_f = qtd_m = qtd_f = 0
for _ in range(n):
    sexo = input("Sexo (M/F): ").strip().upper()
    idade = int(input("Idade: "))
    if sexo == 'M':
        soma_m += idade
        qtd_m += 1
    elif sexo == 'F':
        soma_f += idade
        qtd_f += 1
media_m = soma_m / qtd_m if qtd_m else 0
media_f = soma_f / qtd_f if qtd_f else 0
print("Média idade Masculino:", media_m)
print("Média idade Feminino:", media_f)
```
